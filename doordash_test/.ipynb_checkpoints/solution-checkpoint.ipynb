### 3. Impute missing values 
Use different methods for each column. 

```python
def impute_missing_values(data):

    # market_id: impute with a different category
    data["market_id"].fillna(max(data["market_id"]) + 1, inplace=True)

    # actual_delivery_time: drop all rows with no record
    if "actual_delivery_time" in data.columns:
        data.dropna(subset=["actual_delivery_time"], inplace=True)

    # store_primary_category: replace with "Other"
    data["store_primary_category"].fillna("Other", inplace=True)

    # order_protocol: impute with a different category
    data["order_protocol"].fillna(
        max(data["order_protocol"]) + 1, inplace=True)

    # total_onshift_dashers: impute with median value
    data["total_onshift_dashers"].fillna(np.nanmedian(
        data["total_onshift_dashers"]), inplace=True)

    # total_busy_dashers: impute with median value
    data["total_busy_dashers"].fillna(np.nanmedian(
        data["total_busy_dashers"]), inplace=True)

    # total_outstanding_orders: impute with median value
    data["total_outstanding_orders"].fillna(np.nanmedian(
        data["total_outstanding_orders"]), inplace=True)

    # estimated_store_to_consumer_driving_duration: impute with median value
    data["estimated_store_to_consumer_driving_duration"].fillna(np.nanmedian(
        data["estimated_store_to_consumer_driving_duration"]), inplace=True)

    # Verify that the data contains no missing values (should be False)
    print("Checking if columns contain NaNs:\n")
    for col in list(data.columns):
        print(col, data[col].isnull().values.any())

    return data
```

```python
data = impute_missing_values(data)
data.head()
```

### 4. Add new features to the dataset
Transform the existing features in a way that makes sense. 

```python
def add_features(data):
    # Add total estimated duration
    data["estimated_total_duration"] = data["estimated_order_place_duration"] + \
        data["estimated_store_to_consumer_driving_duration"]

    if "actual_delivery_time" in data.columns:
        # Compute and add total duration
        data["total_duration"] = pd.to_datetime(
            data["actual_delivery_time"]) - pd.to_datetime(data["created_at"])
        data["total_duration"] = data["total_duration"].dt.total_seconds()

    # Delivery time may be affected by traffic conditions, which may depend on month, day of week, and time of day
    # Add this data to the dataset
    # Add month
    data.insert(data.columns.get_loc("store_id"), "created_at_month",
                pd.to_datetime(data["created_at"]).dt.month)
    # Add day of week
    data.insert(data.columns.get_loc("store_id"), "created_at_dayofweek",
                pd.to_datetime(data["created_at"]).dt.dayofweek)
    # Add hour of the day
    data.insert(data.columns.get_loc("store_id"), "created_at_hour",
                pd.to_datetime(data["created_at"]).dt.hour)

    return data


data = add_features(data)
```

### 11. Use XGBoost with decision trees to find the best predictors of total duration
Decision trees have intrinsic feature selection, and running it recursively may help to find the predictors that are less important than the others. Because of the presence of outliers, mean absolute error (MAE) is chosen as a metric of model performance.

```python
# Select columns to use and create a data subset
columns_to_use = ["market_id", "store_id", "store_primary_category", "order_protocol", "total_items", "subtotal",
                  "created_at_month", "created_at_dayofweek", "created_at_hour",
                  "num_distinct_items", "min_item_price", "max_item_price",
                  "total_onshift_dashers", "total_busy_dashers",
                  "total_outstanding_orders", "estimated_order_place_duration",
                  "estimated_store_to_consumer_driving_duration",
                  "estimated_total_duration", "total_duration"]
data_xgb = data[columns_to_use]

# Create arrays with predictors and the target variable
predictors = columns_to_use[:-1]
X = np.array(data_xgb.iloc[:, :-1])
y = np.array(data_xgb["total_duration"]).reshape(-1, 1)
```

```python
# Define Huber(1) loss function 
# Use it instead of the default one because it is close to MAE and close to log-cosh)
def huber_approx_obj(y_true, y_pred):
    d = y_true - y_pred
    h = 1 
    scale = 1 + (d / h) ** 2
    scale_sqrt = np.sqrt(scale)
    grad = d / scale_sqrt
    hess = 1 / scale / scale_sqrt
    return grad, hess

# Define XGBoost hyperparameters
# Hyperparameters are tuned with hyperparameter optimizer(see Appendix 1)
xgb_hyperparameters = {
    "obj": huber_approx_obj,
    "booster": "gbtree",
    "tree_method": "gpu_hist", # Use GPU for computation
    "gpu_id": 0,
    "n_estimators": 1000, # The maximum I could afford
    "max_depth": 8, # Probably needs to be deeper but too long to compute
    "eta": 0.05,
    "gamma": 0}

# Create XGBoost regressor instance
xgb = xgboost.XGBRegressor(**xgb_hyperparameters)
```

```python
# Fit the model to the data 
xgb.fit(X, y)
```

```python
# Get and sort feature importances
feature_imp_list = []
for feature, importance in zip(data.columns, xgb.feature_importances_):  
    temp = [feature, importance * 100]
    feature_imp_list.append(temp)
feature_imp_list.sort(key=lambda x: x[1], reverse=True)

# Print feature importances
print("Feature importances are (descending order): \n")
for feature, importance in feature_imp_list:
    print(f"{feature}: {importance}")
```

```python
# Perform cross-validation to estimate prediction performance
cv = RepeatedKFold(n_splits=10, n_repeats=2, random_state=1234)
scores = cross_val_score(
    xgb, X, y, scoring="neg_mean_absolute_error", cv=cv, n_jobs=-1)

# Compute cross-validation median and median absolute deviation for MAE
mae = [abs(score) for score in scores]
mae_median = np.median(mae)
mae_mad = mad(mae, scale="normal")

# Print the status
print("MAE cross-validation results are: ")
print(mae)
print(f"\nMedian MAE is {mae_median}")
print(f"\nMAE median absolute deviation (scaled) is {mae_mad}")
```

### 13.  Remove the least important features one by one and compare cross-validation MAE
Implement customized recursive feature elemination (RFE) with cross-validation, to select features. 
As less features are selected, the error of the predition (median CV MAE) grows. However, eliminating 9 features do not result in drastic increase in error.

We will proceed with all predictors for the final inference. 

```python
# Subset the predictors
predictors = columns_to_use[:-1]

# Select minimum number of features to stop the process
min_features = 2

# Create a dictionary to store XGBoost cross-validation results
xgb_cv_results = {"predictors_used": [],
                  "predictors_eliminated": [],
                  "mae": [],
                  "mae_median": [],
                  "mae_mad": []}

# Run the cycle while minimum number of predictiors is not reached
while len(predictors) >= min_features:
    xgb_cv_results["predictors_used"].append(predictors.copy())
    print(f"\nUsing {len(predictors)} features")
    print(
        f"Eliminated predictors are: {xgb_cv_results['predictors_eliminated']}")
    print(
        f"Predictors used are: {xgb_cv_results['predictors_used']}")

    # Select data
    data_xgb = data[columns_to_use]
    X = np.array(data_xgb[predictors])
    y = np.array(data_xgb["total_duration"]).reshape(-1, 1)

    # Create a regressor instance
    xgb = xgboost.XGBRegressor(**xgb_hyperparameters)
    xgb.fit(X, y)
    feature_importances = xgb.feature_importances_

    # Extimate K-fold cross-validation metrics
    cv = RepeatedKFold(n_splits=10, n_repeats=2, random_state=1234)
    scores = cross_val_score(
        xgb, X, y, scoring="neg_mean_absolute_error", cv=cv, n_jobs=-1)
    mae = [abs(score) for score in scores]
    mae_median = np.median(mae)
    mae_mad = mad(mae, scale="normal")

    # Update the results
    xgb_cv_results["predictors_eliminated"].append(
        predictors.pop(np.argmin(feature_importances)))
    xgb_cv_results["mae"].append(mae)
    xgb_cv_results["mae_median"].append(mae_median)
    xgb_cv_results["mae_mad"].append(mae_mad)
    
    # Transform to data frame and save the results
    pd.DataFrame(xgb_cv_results).to_csv("cv_results.csv")

    # Print the results
    print(f"MAE are: {mae}")
    print(
        f"MAE median {mae_median} and median absolute deviation (scaled) {mae_mad}")
```

```python
# Plot cross-validation results
xgb_cv_results = pd.DataFrame(xgb_cv_results)
cv_plot = plt.errorbar(list(xgb_cv_results.index.values),
                       xgb_cv_results["mae_median"], yerr=xgb_cv_results["mae_mad"])
cv_plot = plt.title("K-fold cross-validation results for XGBoost")
cv_plot = plt.xlabel("Features eliminated")
cv_plot = plt.ylabel(
    "Cross-validation MAE, median and median absolute deviation (scaled)")
```

### 14. Build XGBoost models to predict confidence interval for the predicted total time
Perform quantile regression with XGBoost. 

We will use regularized quantile loss function with XGBoost, as described here: https://towardsdatascience.com/confidence-intervals-for-xgboost-cac2955a8fde 

```python
# Define custom loss function for quantile regression
def log_cosh_quantile(alpha):
    def _log_cosh_quantile(y_true, y_pred):
        err = y_pred - y_true
        err = np.where(err < 0, alpha * err, (1 - alpha) * err)
        grad = np.tanh(err)
        hess = 1 / np.cosh(err) ** 2
        return grad, hess
    return _log_cosh_quantile

# Define hyperparameters. Leave the same hyperparameters without tuning to save time
# The process in Appendix 1 can be used to finetune the hyperparameters
def xgb_confint_hyperparameters(alpha):
    xgb_confint_hyperparameters = {
        "objective": log_cosh_quantile(alpha),
        "booster": "gbtree",
        "tree_method": "gpu_hist",
        "gpu_id": 0,
        "n_estimators": 1000,
        "max_depth": 8,
        "eta": 0.05,
        "gamma": 0}

    return xgb_confint_hyperparameters


# Define alpha (confidence level for the interval)
alpha = 0.95

# Create the regressors
xgb_lower = xgboost.XGBRegressor(
    **xgb_confint_hyperparameters((1 - alpha) / 2))
xgb_upper = xgboost.XGBRegressor(
    **xgb_confint_hyperparameters((alpha + (1 - alpha) / 2)))
```

```python
# Select data, train the models and predict the confidence intervals
data_xgb = data[columns_to_use]
predictors = columns_to_use[:-1]
X = np.array(data_xgb[predictors])
y = np.array(data_xgb["total_duration"]).reshape(-1, 1)

# Create train/test split
X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.20, random_state=1234)

# Scaling y is necessary to reduce numeric overflow in the loss function
# Use robust scaler 
y_scaler_robust = RobustScaler()
y_train = y_scaler_robust.fit_transform(y_train)

# Train the models
xgb_lower.fit(X_train, y_train)
xgb_upper.fit(X_train, y_train)

# Predict on the test set
y_val_pred_lower = xgb_lower.predict(X_val)
y_val_pred_lower = y_scaler_robust.inverse_transform(
    y_val_pred_lower.reshape(-1, 1))
y_val_pred_upper = xgb_upper.predict(X_val)
y_val_pred_upper = y_scaler_robust.inverse_transform(
    y_val_pred_upper.reshape(-1, 1))
```

```python
# Evaluate the reliability of confidence intervals
count_good_ci = y_val[(y_val >= y_val_pred_lower) &
                      (y_val <= y_val_pred_upper)].shape[0]
count_total_val = len(y_val)
ci_val_prop = count_good_ci / count_total_val * 100
print(
    f"Confidence interval estimated proportion of coverage is: {round(ci_val_prop)} %")
```

```python
# Now train the models on the whole dataset
# Create dataset
data_xgb = data[columns_to_use]
predictors = columns_to_use[:-1]
X = np.array(data_xgb[predictors])
y = np.array(data_xgb["total_duration"]).reshape(-1, 1)

# Initialize the regressors
xgb_lower = xgboost.XGBRegressor(
    **xgb_confint_hyperparameters((1 - alpha) / 2))
xgb_upper = xgboost.XGBRegressor(
    **xgb_confint_hyperparameters((alpha + (1 - alpha) / 2)))

# Use robust scaler to scale y
y_scaler_robust = RobustScaler()
y = y_scaler_robust.fit_transform(y)

# Train the models
xgb_lower.fit(X, y)
xgb_upper.fit(X, y)
```

```python
# Predict the test data set
# Load the test data
test_data = pd.read_csv("test_results_xgboost.csv")
test_data.drop(test_data.columns[0], axis=1, inplace=True)

# Create test data subset
test_data_predictors = test_data[predictors]
X = np.array(test_data_predictors)

# Predict the confidence bounds
y_test_pred_lower = xgb_lower.predict(X)
y_test_pred_lower = y_scaler_robust.inverse_transform(
    y_test_pred_lower.reshape(-1, 1))
y_test_pred_lower = y_test_pred_lower.astype(int)
y_test_pred_upper = xgb_upper.predict(X)
y_test_pred_upper = y_scaler_robust.inverse_transform(
    y_test_pred_upper.reshape(-1, 1))
y_test_pred_upper = y_test_pred_upper.astype(int)

# Save the prediction in the file
test_data["pred_lower_ci"] = y_test_pred_lower
test_data["pred_upper_ci"] = y_test_pred_upper
test_data.to_csv("test_results_xgboost_with_confint.csv")
```

### Appendix 1. Perform tuning of hyperparameters of XGBoost using optimizer of Lipschitz functions (dlib implementation)
This process takes a lot of time and has to be done overnight.
The optimizer has been run twice. The first trial (**Attachment 2**) included tree depth, and converged at optimal tree depth of 8. 
The second run involved fine-tuning of eta and gamma with fixed depth of 8. This has been done because this optimizer works better with continuous variables. 

```python
# Link to the paper: https://arxiv.org/abs/1703.02628
# Optimize the hyperparameters of the XGBoost using dlib

# Select data
data_xgb = data[columns_to_use]
predictors = columns_to_use[:-1]
X = np.array(data_xgb[predictors])
y = np.array(data_xgb["total_duration"]).reshape(-1, 1)

# Results
test_results = {"hyperparameters": [],
                "mae_median": []
                }

# Tune maximum depth, learning rate and gamma parameters
def function_to_optimize(eta, gamma): # + max_depth
    # Transform the input parameters
    # max_depth = int(max_depth)
    print(f"\nProbing eta: {eta}, gamma: {gamma}")

    # Set the number of trees as high as I can afford within my resources and time frame
    xgb_hyperparameters = {"obj": huber_approx_obj,
                           "booster": "gbtree",
                           "tree_method": "gpu_hist",
                           "gpu_id": 0,
                           "n_estimators": 1000,
                           "max_depth": 8,
                           "eta": eta,
                           "gamma": gamma}
    xgb = xgboost.XGBRegressor(**xgb_hyperparameters)
    cv = RepeatedKFold(n_splits=10, n_repeats=2, random_state=1234)
    scores = cross_val_score(
        xgb, X, y, scoring="neg_mean_absolute_error", cv=cv, n_jobs=-1)
    mae = [abs(score) for score in scores]
    mae_median = np.median(mae)

    # Save the results in .csv file
    test_results["hyperparameters"].append([eta, gamma])
    test_results["mae_median"].append(mae_median)
    pd.DataFrame(test_results).to_csv("hyperparam_opt_results.csv")

    # Print the target
    print(f"Target is: {mae_median}")

    return mae_median


# Perform optimization of hyperparameters
dlib.find_min_global(function_to_optimize, [0.001, 0], [0.5, 50], 200)
```
